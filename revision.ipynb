```python
import pandas as pd
import mysql.connector
```

```python
conn=mysql.connector.connect(
    host='localhost',
    user='root',
    password='',
    database='revision'
    )
```

```python
cursor=conn.cursor()
```

```python
file=pd.read_csv('C:/Users/user/Downloads/Education_Facilities.csv')
file
```

```python
school_loca=file[['PROVINCE','DISTRICT','SECTOR','CELL']]
school_loca=school_loca.rename(columns={'PROVINCE':'l_province','DISTRICT':'l_district','SECTOR':'l_sector','CELL':'l_cell'})
school_loca
```

```python
insert_query=f"INSERT INTO school_loc({','.join(school_loca.columns)})VALUES({','.join(['%s']*len(school_loca.columns))})"
for _,row in school_loca.iterrows():
    values=tuple(row)
    cursor.execute(insert_query,values)
conn.commit()
```

```python
# Execute SQL query
cursor.execute("SELECT * FROM school_loc")
# Fetch data and column names
rows = cursor.fetchall()
column_names = [i[0] for i in cursor.description]

# Create DataFrame with column names
school_loc= pd.DataFrame(rows, columns=column_names)

# Display results
school_loc
```
