```python
query = """
SELECT 
    ROW_NUMBER() OVER (ORDER BY c.customer_id) AS customer_id_SC,
    c.customer_id,
    a.address_id,
    a.country_id,
    ca.status_id,
    c.first_name,
    c.last_name,
    c.email,
    cn.country_name,
    a.street_name,
    a.city,
    as_status.address_status,
    'GRAVITY_BOOKS' AS source_system_code,
    GETDATE() AS start_date,
    1 AS is_current
FROM customer c
JOIN customer_address ca ON c.customer_id = ca.customer_id
JOIN address a ON ca.address_id = a.address_id
JOIN country cn ON a.country_id = cn.country_id
JOIN address_status as_status ON ca.status_id = as_status.status_id;
"""

# ✅ Extract Data into DataFrame
df = pd.read_sql(query, conn)

# ✅ Display Extracted Data
df
```

```python
# SQL Query to extract shipping method data with all required columns
shipping_query = """
SELECT 
    ROW_NUMBER() OVER (ORDER BY sm.method_id) AS method_id_SC,
    sm.method_id,
    sm.method_name,
    sm.cost,
    'GRAVITY_BOOKS' AS source_system_code,
    GETDATE() AS start_date,
    NULL AS end_date,    -- Added end_date column initialized as NULL
    1 AS is_current
FROM shipping_method sm;
"""

# Extract Data into DataFrame
dim_order_df = pd.read_sql(shipping_query, conn)

# Display Extracted Data
print("Extracted Shipping Method Data:")
display(dim_order_df.head())

# Verify column names and structure
print("\nDataFrame Columns:")
print(dim_order_df.columns.tolist())
```

```python
# SQL Query to generate date dimension data
dim_date_query = """
WITH DateRange AS (
    SELECT DATEADD(DAY, number, '2010-01-01') AS date_value
    FROM master.dbo.spt_values
    WHERE type = 'P' 
    AND number BETWEEN 0 AND DATEDIFF(DAY, '2010-01-01', GETDATE())
)
SELECT 
    ROW_NUMBER() OVER (ORDER BY date_value) AS date_id_SC,
    YEAR(date_value) AS year,
    MONTH(date_value) AS month,
    DAY(date_value) AS day,
    date_value AS start_date,
    NULL AS end_date,
    'GRAVITY_BOOKS' AS source_system_code,
    1 AS is_current
FROM DateRange
WHERE date_value <= GETDATE();
"""

# Extract Data into DataFrame
dim_date_df = pd.read_sql(dim_date_query, conn)

# Display Extracted Data
print("Extracted Date Dimension Data:")
display(dim_date_df.head())

# Verify column names and structure
print("\nDataFrame Columns:")
print(dim_date_df.columns.tolist())

# Show date range summary
print("\nDate Range Coverage:")
print(f"From: {dim_date_df['start_date'].min().date()}")
print(f"To: {dim_date_df['start_date'].max().date()}")
print(f"Total days: {len(dim_date_df)}")
```

```python
# SQL Query to extract fact order data with order_status
fact_order_query = """
SELECT 
    ROW_NUMBER() OVER (ORDER BY co.order_id, ol.line_id) AS order_id_SC,
    ol.line_id,
    co.order_id,
    oh.history_id,
    ol.book_id,                      -- Source system book_id
    co.customer_id,                  -- Source system customer_id
    co.shipping_method_id,           -- Source system method_id
    CONVERT(DATE, co.order_date) AS order_date,  -- Source date
    os.status_id,
    os.status_value AS order_status,  -- Added order_status column
    ol.price,
    sm.cost AS shipping_cost,
    'GRAVITY_BOOKS' AS source_system_code,
    GETDATE() AS created_at
FROM cust_order co
JOIN order_line ol ON co.order_id = ol.order_id
JOIN order_history oh ON co.order_id = oh.order_id
JOIN order_status os ON oh.status_id = os.status_id
JOIN shipping_method sm ON co.shipping_method_id = sm.method_id;
"""

# Extract Data into DataFrame
fact_order_df = pd.read_sql(fact_order_query, conn)

# Display Extracted Data
print("Extracted Fact Order Data with Status:")
display(fact_order_df.head())

# Verify column names and structure
print("\nDataFrame Columns:")
print(fact_order_df.columns.tolist())

# Show status value counts
print("\nOrder Status Distribution:")
print(fact_order_df['order_status'].value_counts())
```
